## **What is a Decision Tree ?**

Decision Tree, juga dikenal dengan `Classification and Regression Trees (CART)` , adalah supervised machine learning algorithm untuk classification dan regression problems. Decision tree dibangun seperti model struktur flowchart, dimana decision dibuat dari sekumpulan `"if-then-else"

1.JPG` statements.

image.png

Berdasarkan diagram diatas, terminologi nya yaitu :
- **Root node** = Node pertama dalam tree.
- **Internal nodes** = Ini adalah decision nodes.
- **Branches** = Panahnya menghubungkan nodes. Merepresentasikan langkah decision-making
- **Leaf nodes** = Nodes ini merepresentasikan semua kemungkinan outcomes dari dataset.

Decision Tree memungkinkan continuous data splitting berdasarkan parameter tertentu hingga final decision tercapai. Mereka memulai dengan single node yang kemudian branches menjadi possible outcomes. Possible outcomes dipecah menjadi additional nodes, branching menjadi hasil yang lebih mungkin, sehingga membentuk struktur tree.

Misalnya, saat merencanakan sebuah pertandingan, kami mungkin ingin memutuskan apakah pertandingan akan berlangsung bergantung pada data parameter seperti cuaca, kelembapan, dan angin. Tree yang dihasilkan akan menjadi seperti ini:

## **Types of Decision Trees**

Decision Tree dikategorikan berdasarkan jenis variabel target. Mereka dapat terdiri dari dua jenis:

- **Continuous Variable Decision tree :** Decision Tree yang variabel targetnya continuous. Misalnya, kita dapat membuat decision tree untuk menentukan usia seseorang berdasarkan DNA methylation levels.
- **Categorical Variable Decision tree :** Decision Tree yang variabel targetnya bersifat categorical. Suatu kategori dapat berupa ya atau tidak, artinya decision tersebut hanya termasuk dalam satu kategori untuk setiap tahapan.

## **Decision Tree Algorithms**

Algoritma yang digunakan dalam Decision Tree juga didasarkan pada jenis variabel target. Itu termasuk :

- **ID3 (Iterative Dichotomiser)** = Algoritma ini menggunakan **entropy** dan **information gain** untuk menentukan split points. Ini menggunakan pendekatan pencarian serakah untuk memilih atribut terbaik untuk membagi dataset pada setiap iterasi.
- **C4.5 Algorithm** = Ini menggunakan **informtion gain** untuk mengevaluasi split points. Ini memperluas Quinlans earlier ID3 algorithm dan sering disebut sebagai statistical classifier.
- **Classification and Regression Trees (CART)** = Ini adalah algoritma yang paling mapan dan banyak digunakan.
>- Algoritma ini adalah non-parametric decision tree learning technique.
>- Algortitma ini menghasilkan classification or regression trees berdasarkan apakah dependent variable bersifat categorical atau numerical.
>- Menggunakan `Gini impurity` untuk memutuskan split points.
>- Kita juga dapat menggunakan algoritma ini untuk mendeteksi anomali dan outlier.

## **Splitting criteria for Decision Trees**

Ide utama dari decision tree adalah untuk mengidentifikasi features yang berisi informasi yang memadai tentang fitur **target** dan kemudian membagi dataset beserta nilainya. Pencarian atribut yang paling informatif membuat decision tree hingga kita mendapatkan pure leaf nodes.

Ada berbagai kriteria yang digunakan oleh decision tree untuk memilih atribut terbaik untuk pemisahan pada setiap node. Ini termasuk :
- Entropy
- Information gain
- Gini impurity

## **Assumptions of Decision Trees**

Decision Tree tidak membuat asumsi tentang training data atau prediction residues karena menggunakan pendekatan non-statistik. Namun, ada beberapa asumsi yang kami buat saat membuat decision tree :
- Saat kita mulai membangun tree, seluruh training set diasumsikan sebagai root.
- Feature value pilihan bersifat categorical. Jika kita memiliki nilai continuous, nilai tersebut akan didiskritisasi sebelum model dibuat.
- Records didistribusikan secara rekursif berdasarkan nilai atribut.
- Urutan penempatan atribut sebagai root atau node internal dilakukan dengan menggunakan pendekatan statistik.

## **What is a Random Forest?**

Sangat mudah untuk memahami dan memahami Random Forest Algorithm. Namun, terkadang menghasilkan single tree saja tidak cukup untuk memberikan hasil yang efektif. Di sinilah random forest beraksi.

Random Forest adalah ensemble method yang disebut Bootstrap Aggregation atau bagging yang menggunakan banyak decision tree untuk mengambil keputusan. Seperti namanya, ini sebenarnya adalah forest dari decision tree.

Disebut "random" forest karena:

- Secara random mengambil sampel training data sample untuk membuat tree.
- Menghasilkan random subset of features untuk menghitung output untuk splitting node.

## **General differences between a Decision Tree and Random Forest**

Decision Trees dan Random Forests mempunyai perbedaan yang signifikan. Di bawah ini memberikan perbandingan berdampingan berdasarkan berbagai isu.

- **Decision Tree**
>- Tidak butuh waktu yang lebih untuk menjalankannya

- **Random Forest**
>- Butuh waktu lebih untuk menjalankannya

- Mari kita beri label pada kolom `quality` dengan label buruk dan baik dimana nilai kurang dari 6 dianggap buruk, dan nilai di atas 6 adalah kualitas baik.

```python
# label the quality column

bins = (2, 6, 8) # 2-6 = kualitas buruk, 6-8 = kualitas bagus
groups = ['buruk', 'bagus']
df['quality'] = pd.cut(df['quality'], bins = bins, labels = groups)
df.head(10)
```

### **Splitting Data**

Pertama-tama mari kita bagi dataset menjadi independent dan dependent variable. Feature `quality` akan berfungsi sebagai traget variable.

```python
X = df.drop('quality', axis=1)
y = df['quality']
```

- Split the dataset into training and test set.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

### **Creating Model**

Sekarang saatnya membangun decision tree. Scikit-learn menyediakan class sklearn.tree.DecisionTreeClassifier. Kami akan mempertimbangkan beberapa parameter berikut:

- **criterion** = mengukur kualitas splitting. Kriteria yang didukung adalah `Gini` dan `Entropy` untuk information gain.
- **splitter** = strategi untuk memilih splitting pada setiap node. Strategi `'best'` dan `'random'`.
- **random_state** = mengontrol keacakan splitting.
- **min_samples_split** = Jumlah minimum sampel yang diperlukan untuk memisahkan node internal.
- **min_samples_leaf** = Jumlah minimum sampel yang diperlukan untuk berada pada leaf node.
- **max_leaf_node** = Membuat pohon dengan `max_leaf_nodes` dengan cara terbaik-pertama. Node terbaik didefinisikan sebagairelative reduction in impurity.  Jika None, maka jumlah leaf nodes tidak terbatas.

**Langkah-langkahnya :**
- **import** DecisionTreeClassifier class.
- **Instantiate** classifier model.
- **Fit** model nya.
- **Predict**
- **Evaluate** performance model.

```python
clf = DecisionTreeClassifier(random_state=42) # instantiate the model
clf = clf.fit(X_train, y_train) # fit the model
```

```python
# making predictions
y_pred = clf.predict(X_test) # Test the prediction accurracy of the model
result = pd.DataFrame({'Actual' : y_test, 'Predicted' : y_pred})
pd.concat([result.head(), result.tail()]) # display df of acutal and predicted(head and tail)
```

### **Hyperparameter tuning on Decision Tree Classifier**

Kami akan mengimpor `GridSearchCV` class dari sklearn dan kemudian meneruskan nilai hyperparameter yang telah ditentukan ke dalamnya.

```python
from sklearn.model_selection import GridSearchCV

# defining parameters
param_grid = {
    "criterion":("gini", "entropy"),
    "splitter":("best", "random"),
    "max_depth":(list(range(1, 15))),
    "min_samples_split":[2, 3, 4, 6, 8],
    "min_samples_leaf":list(range(1, 15)),
}
# training model on the define params with GridSearchCV
clf_tree = DecisionTreeClassifier()
tree_cv = GridSearchCV(clf_tree, param_grid,scoring="accuracy",
                       n_jobs=-1, verbose=1, cv=3)

tree_cv.fit(X_train, y_train)
predictions = tree_cv.predict(X_test)
optimal_params = tree_cv.best_params_
print(f"\nOptimal paramters: ({optimal_params})")

print("\n====BEST MODEL SCORE====")
print('Best model score: ', tree_cv.best_score_)

print("\n====CLASSIFICATION REPORT====")
target_names = ['Bad Quality', 'Good Quality']
pd.DataFrame(classification_report(y_test,
                                   predictions,
                                   target_names=target_names,
                                   output_dict=True))
```

Jadi, tree yang paling akurat harus memiliki nilai parameter optimal di atas.

### **Pruning Decision Trees**

Dalam kebanyakan kasus, decision trees cenderung mengalami overfitting. Decision Trees akan overfit ketika diizinkan untuk splitting menjadi node hingga semua leaf pure atau hingga semua leaf berisi kurang dari `min_samples_split` sampel. Artinya, membiarkannya mencapai `max-depth`.

Membiarkan decision tree mencapai maximum depth akan menghasilkan complex tree, seperti pada contoh kita di atas.

Pruning dapat diklasifikasikan menjadi:

**1. Pre-pruning**
- Pre-pruning juga bisa disebut "early stopping". Di sini kami mencoba menghentikan tree agar tidak splitting hingga `maximum depth`. Ingat:
>- Decision Tree yang kami buat di atas complex dan tidak mudah dijelaskan.
>- Kami melakukan hyperparameter tuning, di mana kami mendapatkan nilai parameter optimal.

Untuk menyederhanakan model, kita dapat melatihnya kembali dengan nilai dari hyperparameter tuning, mendapatkan accuracynya, dan memvisualisasikan tree-nya lagi.

```python
clf = DecisionTreeClassifier(criterion='gini',
                             max_depth=4,
                             splitter='best',
                             min_samples_leaf=8,
                             min_samples_split=4,
                             random_state=42
                            )

clf = clf.fit(X_train, y_train)
print('Testing Set Evaluation Accuracy: ',
      accuracy_score(y_test,clf.predict(X_test)))
```

Anda dapat melihat bahwa model accuracy kami mengalami sedikit imporovement. Mari kita visualisasikan :

```python
# visualizing
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data, feature_names=features, filled=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

Kita juga dapat mengamati bahwa aturannya telah dikurangi, dan modelnya sedikit mudah untuk diinterpretasikan.

**2. Post-pruning**

Dalam post-pruning, kami membiarkan tree berlatih dengan sempurna di training set dan kemudian prune. `DecisionTreeClassifier` memiliki parameter bernama `ccp_alpha`. Parameter ini disebut `cost-complexity parameter`.

Umumnya, the cost-complexity printing recursively secara rekursif menemukan node dengan "weakest link". Weakest link ditandai dengan effective alpha, dimana node dengan smallest effective alpha d-prune terlebih dahulu.

Mari kita lihat bagaimana hal itu dilakukan.

```python
path = clf.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities
```

Untuk mendapatkan gambaran tentang nilai parameter `ccp_alpha` yang layak, kami menggunakan `cost_complexity_pruning_path` yang memberi kami alpha dan total leaf impurities yang sesuai pada setiap langkah proses pruning.

ccp_alphas variable stores the weak points with respect to each node in the form of a list.

Variabel `ccp_alphas` menyimpan weak points terhadap setiap node dalam bentuk daftar.

```python
ccp_alphas
```

Kami akan melatih decision tree model menggunakan nilai alfa ini.

```python
# grid search to find the optimal value for alpha from the ccp_alphas list

clf_tree = DecisionTreeClassifier(random_state=42)
ccp_alpha_grid_search = GridSearchCV(clf_tree, param_grid=({"ccp_alpha":[alpha for alpha in ccp_alphas]}))

ccp_alpha_grid_search.fit(X_train, y_train) # # train model
# find the best parameter value
print('Best parameter:' , ccp_alpha_grid_search.best_params_)
```

Kami melihat bahwa decision tree terbaik akan dihasilkan oleh `ccp_alpha` dengan nilai 0,009017930023689974.

# **Building a Random Forest model with Sklearn**

Random Forest cocok dengan beberapa decision tree classifier pada berbagai sub-sampel dataset. Ini menggunakan rata-rata untuk meningkatkan predictive accuracy dan control over-fitting. Random forest mencapai higher classification accuracy daripada decision tree.

- **n_estimators** = Jumlah tree pada forest.
- **criterion** = Berfungsi untuk mengukur kualitas suatu split. Kriteria yang didukung adalah `gini` untuk **Gini Impurity** dan `entropi` untuk **Information Gain**.
-  **max_depth** = Maximum depth dari tree. Jika tidak Ada, maka node diperluas hingga semua leaves(daun) murni atau hingga semua leaves(daun) berisi kurang dari `min_samples_split` sampel.
- **min_samples_split** = Jumlah minimum sampel yang diperlukan untuk split an internal node.
- **min_samples_leaf** = Jumlah minimum sampel yang diperlukan untuk berada pada leaf node. Split point pada depth berapa pun hanya akan dipertimbangkan jika leaves(daun) setidaknya `min_samples_leaf` training sample di masing-masing left and right branches. Hal ini mungkin mempunyai efek memperhalus model, terutama dalam regression.
- **min_weight_fraction_leaf** = The minimum weighted fraction dari sum total of weights (dari semua sampel input) yang diperlukan untuk berada pada leaf node. Sampel memiliki weight yang sama jika sample_weight tidak disediakan.
- **max_features** = Jumlah features yang perlu dipertimbangkan saat mencari best splitting.
- **max_leaf_nodes** = Grow sebuah tre dengan max_leaf_nodes dengan cara best-first. Best nodes didefinisikan sebagai relative reduction in impurity. Jika Tidak Ada makaunlimited number of leaf nodes.
- **min_impurity_decrease** = Sebuah node akan split jika split ini menyebabkan the impurity lebih besar atau sama dengan nilai ini.
- **min_impurity_split** = Threshold penghentian dini pertumbuhan tree. Sebuah node akan split jika impurity berada di atas threshold, jika tidak maka ia adalah leaf.
- **bootstrap** = Apakah sampel bootstrap digunakan saat membuat tree. Jika False, seluruh dataset digunakan untuk membangun setiap tree.
- **oob_score** = Apakah akan menggunakan sampel siap pakai untuk memperkirakan generalization accuracy.

Sebelum kita melanjutkan, mari kita buat sebuah fungsi untuk menghindari terlalu banyak pengulangan.

```python
def eval_model_score(clf, X_train, y_train, X_test, y_test):
    y_pred = clf.predict(X_test)
    print('Testing Set Evaluation Accuracy: ',accuracy_score(y_test,y_pred))
    print("\n====CONFUSION MATRIX====\n")
    print(confusion_matrix(y_test, y_pred)) # generate confusion matrix
    print("\n====CLASSIFICATION REPORT====")
    target_names = ['Bad Quality', 'Good Quality']
    clf_report = pd.DataFrame(classification_report(y_test, y_pred,
                                                    target_names=target_names,
                                                    output_dict=True))
    print(clf_report)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier(random_state=42)
rf_clf.fit(X_train, y_train)
eval_model_score(rf_clf, X_train, y_train, X_test, y_test)
```

Kita dapat langsung mengamati bahwa random forest classification memberikan accuracy lebih dibandingkan decision tree clasification. Nanti, kami akan menjelaskan mengapa random forest memiliki kinerja lebih baik.

## **Hyperparameter tuning on Random Forest Classifier**

Kami dapat menemukan parameter terbaik untuk meningkatkan kinerja model kami dengan menggunakan `sklearn.model_selection.RandomizedSearchCV` class.

Perbedaan utama antara `RandomSearchCV` dan `GridSearchCV` adalah meskipun `GridSearchCV` menguji semua parameter dan memilih yang terbaik, `RandomSearchCV` tidak menguji semua parameter melainkan mencari secara acak.

`RandomSearchCV` juga memiliki parameter `n_iter` yang membantu menentukan jumlah nilai parameter yang ingin kita uji.

```python
from sklearn.model_selection import RandomizedSearchCV

# Provide hyperparameter grid for a random search
n_estimators = [int(x) for x in np.linspace(start=200, stop=2000, num=10)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(10, 120, num=12)]
max_depth.append(None)
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]
bootstrap = [True, False]

random_grid = {'n_estimators': n_estimators, 'max_features': max_features,
               'max_depth': max_depth, 'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf, 'bootstrap': bootstrap}

rf_cv = RandomizedSearchCV(estimator=rf_clf, scoring='accuracy',
                           param_distributions=random_grid,
                           n_iter=50, cv=5, verbose=2,
                           random_state=42, n_jobs=-1
                          )

rf_cv.fit(X_train, y_train)
rf_best_params = rf_cv.best_params_
print(f"Best paramters: {rf_best_params})")

eval_model_score(rf_clf, X_train, y_train, X_test, y_test)
```

Train the model with the tuned parameters.

```python
rf_clf = RandomForestClassifier(**rf_best_params) # pass the best parameters as **kwargs
rf_clf.fit(X_train, y_train)
eval_model_score(rf_clf, X_train, y_train, X_test, y_test)
```

## **Why Random Forest performed better than Decision Tree**
